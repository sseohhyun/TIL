```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# 1. 펭귄 데이터 불러오기
df = sns.load_dataset("penguins")

# 결측치 제거 (body_mass_g에 NaN이 포함되어 있음)
df = df.dropna(subset=["body_mass_g"])

df
```

```python
# 2. 모집단 평균
population_mean = df['body_mass_g'].mean()

population_mean
```

```python
# 3. 표본 크기 n=30으로 100번 임의 추출하여 각 표본 평균을 계산하세요.
sample_means = []
for i in range(100):
    sample = np.random.choice(df['body_mass_g'], size=30, replace=True)   # TODO: body_mass_g 컬럼에서 임의로 30개 추출
    sample_means.append(sample.mean())  # TODO: 표본 평균 저장

# 4. 모집단 평균과 표본 평균 분포를 시각화하세요.
plt.hist(sample_means, bins=20, alpha=0.7, label="Sample Means")
plt.axvline(population_mean, color="red", linestyle="--", label="Population Mean")
plt.legend()
plt.show()
```
