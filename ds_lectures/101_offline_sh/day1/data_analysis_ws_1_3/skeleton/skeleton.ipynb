```python
import pandas as pd  # Pandas 라이브러리 불러오기
# =============================================================================
# quantile() 함수 설명
# quantile() 함수는 데이터에서 특정 분위수(퍼센트) 값을 구하는 데 사용됩니다.
# 예를 들어, quantile(0.90)은 상위 10%에 해당하는 값의 기준을 의미합니다.

# 기본 문법:
# df["컬럼명"].quantile(퍼센트)
# =============================================================================

# 1. Titanic 데이터셋 불러오기 (파일 경로는 사용자 환경에 맞게 변경)
df = pd.read_csv("C:/Users/SSAFY/Desktop/TIL/ds_lectures/101_offline_sh/day1/data_analysis_ws_1_3/data/titanic.csv")  # Titanic 데이터셋을 불러옴
```

```python
# 2. 데이터프레임 기본 정보 확인
print("데이터프레임 기본 정보:")
print(df.info())  # 데이터 구조 및 결측치 개수 확인
```

```python
# 3. 생존자 데이터 필터링
# Survived 값이 1인 승객만 필터링 (생존한 승객)
survived_passengers = df[df["Survived"] == 1]
print("\n생존한 승객 데이터 (상위 5개):")
survived_passengers.head()  # 생존자 상위 5개 데이터 확인
```

```python
# 4. 사망자 데이터 필터링
# Survived 값이 0인 승객만 필터링 (사망한 승객)
deceased_passengers = df[df["Survived"] == 0]
print("\n사망한 승객 데이터 (상위 5개):")
deceased_passengers.head()  # 사망자 상위 5개 데이터 확인
```

```python
# 5. 요금(Fare) 결측치 확인 및 처리
# Fare 컬럼의 결측치 개수 확인
print("\n요금(Fare) 컬럼 결측치 개수:")
print(df["Fare"].isna().sum())  # 결측치 개수 출력

# 결측치가 있으면 0으로 채우기 (비행기 탑승 기록이 없는 경우를 고려)
df["Fare"] = df["Fare"].fillna(0) # 결측치를 0으로 대체하는 함수 입력
```

```python
# 6. 고액 요금 승객 필터링
# 요금이 상위 10%에 해당하는 값 찾기
fare_threshold = df["Fare"].quantile(0.9)  # 상위 10%의 기준값 설정

# 고액 요금 승객 필터링 (요금이 fare_threshold 이상인 승객)
high_fare_passengers = df[df["Fare"] >= fare_threshold]
print(f"\n고액 요금 승객 데이터 (상위 5개) - 기준 요금: {fare_threshold:.2f}")
high_fare_passengers.head()  # 고액 요금 승객 상위 5개 데이터 확인
```

```python
# 7. 생존한 고액 요금 승객 필터링
# 생존자 중에서 요금이 fare_threshold 이상인 승객
high_fare_survivors = df[(df['Survived'] == 1) & (df["Fare"] >= fare_threshold)]
print("\n생존한 고액 요금 승객 데이터 (상위 5개):")
high_fare_survivors.head() # 생존한 고액 요금 승객 상위 5개 데이터 확인
```

```python
# 8. 사망한 고액 요금 승객 필터링
# 사망자 중에서 요금이 fare_threshold 이상인 승객
high_fare_deceased = df[(df['Survived'] == 0) & (df["Fare"] >= fare_threshold)]
print("\n사망한 고액 요금 승객 데이터 (상위 5개):")
high_fare_deceased.head()  # 사망한 고액 요금 승객 상위 5개 데이터 확인
```
