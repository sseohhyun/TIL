```python
# 시나리오 설명
'''
Assignment 1에서 생성된 고객 알림 데이터에는 고객 ID와 alert 여부 정보만 포함되어 있습니다.
김싸피가 속한 마케팅 팀은 각 국가별로 알림 대상 고객을 별도 파일로 분리하여 마케팅 담당자에게 전달하고자 합니다.
이를 위해 고객 원본 데이터에서 지역 정보를 병합하고, alert == "yes"인 고객을 국가 단위로 분할하여 저장하세요.
'''
```

```python
# 요구 사항
'''
assignment1/output/alert_lifetime_summary.csv 파일을 불러옵니다.
CustomerLoyaltyProgram.csv 파일도 data 폴더에 위치합니다.
Loyalty# 기준으로 지역 정보를 병합합니다,
alert == "yes" 조건으로 필터링합니다.
국가(Country) 기준으로 그룹핑합니다.
../output/reminder_by_region/ 폴더에 국가별 CSV 저장합니다.
'''
```

```python
# 1. 라이브러리 불러오기
import pandas as pd
import os
```

```python
# 2. 데이터 불러오기
# Assignment 1에서 저장된 알림 대상 고객 데이터를 불러옵니다
summary_df = pd.read_csv("../output/alert_lifetime_summary.csv")

# 고객 지역 정보 (CustomerLoyaltyProgram.csv)를 불러옵니다
region_df = pd.read_csv("../data/CustomerLoyaltyProgram.csv")
```

```python
region_df.head()
```

```python
summary_df.head()
```

```python
# 3. 지역 병합 정보 생성
# Loyalty#를 기준으로 병합할 지역 정보 테이블을 생성합니다
region_info = region_df[["Loyalty#", "Country", "Province or State", "City"]].drop_duplicates()

# 고객 요약 정보와 병합하여 지역 정보를 포함한 데이터프레임을 생성하세요
merged_df = summary_df.merge(region_info, on="Loyalty#", how="inner")
```

```python
# 4. 알림 대상 필터링
# 'alert' 값이 'yes'인 데이터만 필터링하세요
alert_df = merged_df[merged_df["alert"] == "yes"].copy()
```

```python
# 5. 결과 저장 폴더 만들기
# reminder_by_region 경로를 생성하세요
output_dir = "../output/reminder_by_region"
os.makedirs(output_dir, exist_ok=True)
```

```python
# 6. 국가별로 알림 고객 분리 저장
# Country를 기준으로 그룹핑하여 파일을 나눠 저장합니다
for country, group in alert_df.groupby("Country"):
    filename = f"reminder_{country.replace(' ', '_')}.csv"
    save_path = os.path.join(output_dir, filename)
    group.to_csv(save_path, index=False)
    print(f"[{country}] 저장 완료 - {len(group)}명")
```
