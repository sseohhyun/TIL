```python
import pandas as pd

df = pd.read_csv("../data/predictions.csv")
df.head(5)
```

```python
print(df["predictions"].value_counts())
```

```python
df.info()
X_pred = df.drop(columns=["predictions"])
```

```python
# Tracking Server 설정
import mlflow

mlflow.set_tracking_uri(uri="http://127.0.0.1:5000")
mlflow.set_experiment("creditcard_experiment_6_4")

MODEL_URI = "runs:/5240fefe24da433ab63ce23c7186ada9/model" 

loaded_model = mlflow.pyfunc.load_model(MODEL_URI)
new_predictions = loaded_model.predict(X_pred)
df["new_predictions"] = new_predictions
```
